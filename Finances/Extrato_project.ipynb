```python
Extract = pd.DataFrame(data)
print(Extract.shape)
Extract.head
```

```python
drop_cols = [2,3,6]
Extract_drop_rows = Extract.drop([0,1,2,3,4,5,6])
Extract_drop = Extract_drop_rows.drop(Extract.columns[drop_cols], axis=1)
Extract_drop.head(20)
```

```python
Extract_names = Extract_drop.rename(index=str, columns={0: "Data",1: "Descricao",4:"Receita",5:"Despesa"})
```

```python
Credito_mil =  Extract_names['Receita'].str.replace('.','')
Debito_mil = Extract_names['Despesa'].str.replace('.','')
del Extract_names['Receita']
del Extract_names['Despesa']
Extract_names['Receita'] = Credito_mil
Extract_names['Despesa'] = Debito_mil
```

```python
Extract_names
```

```python
Credito =  Extract_names['Receita'].str.replace(',','.')
Debito = Extract_names['Despesa'].str.replace(',','.')
del Extract_names['Receita']
del Extract_names['Despesa']
Extract_names['Receita'] = Credito
Extract_names['Despesa'] = Debito
```

```python
Credito_0 =  Extract_names['Receita'].str.replace('^\s','0.00')
Debito_0 = Extract_names['Despesa'].str.replace('^\s','0.00')
del Extract_names['Receita']
del Extract_names['Despesa']
Extract_names['Receita'] = Credito_0
Extract_names['Despesa'] = Debito_0
```

```python
Extract_names
```

```python
Credito_float = Extract_names['Receita'].astype(float)
Debito_float = Extract_names['Despesa'].astype(float)
del Extract_names['Receita']
del Extract_names['Despesa']
Extract_names['Receita'] = Credito_float
Extract_names['Despesa'] = Debito_float
```

```python
Extract_names.loc['12','Data']
```

```python
Extract_names['Categoria'] = np.nan
Extract_names['Banco'] = 'Santander'
Extract_names.head(20)
```

```python
Extrato_santander = Extract_names
```

## writing

```python
file = 'financas.xlsx'
xlsx_file = pd.ExcelFile(file) 
xlsx_file.sheet_names
```

```python
df = xlsx_file.parse('input',header= None) 
df_NoHeader = df[2:] 
old_data = df_NoHeader 
old_data_lab = old_data.rename(index=str, columns={0: "Data",1: "Descricao",2:"Receita",3:"Despesa",4:"Categoria",5:"Banco"})
old_data_lab.head(50)
```

```python
new_data = old_data_lab.append(Extrato_santander)
new_data.head(50)
```

```python
import xlwings as xw

wb = xw.Book('financas.xlsx')
ws = wb.sheets['input']

ws.range('A2').options(index=False, header=True).value = new_data
wb = xw.Book('financas_update.xlsx')

xw.apps[0].quit()
```
